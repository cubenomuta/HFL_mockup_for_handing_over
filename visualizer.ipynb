# OrganAMNIST

```python
# 必要なライブラリのインポート
import matplotlib.pyplot as plt
from collections import Counter
from load_medmnist import load_organmnist_data  # データ読み込み関数をインポート


# データセットの準備関数（例: 訓練データ、検証データ、テストデータ）
def load_and_count_labels(data_flag, split):
    dataset = load_organmnist_data(data_flag=data_flag, split=split)
    labels = [int(label[0]) for _, label in dataset]
    label_counts = Counter(labels)
    return label_counts

# 訓練データ、検証データ、テストデータのラベル分布の取得
train_counts = load_and_count_labels(data_flag='organamnist', split='train')
val_counts = load_and_count_labels(data_flag='organamnist', split='val')
test_counts = load_and_count_labels(data_flag='organamnist', split='test')

# ラベル分布の可視化関数
def plot_label_distribution(label_counts, title):
    plt.bar(label_counts.keys(), label_counts.values())
    plt.xlabel("Label")
    plt.ylim(0, 6500)
    plt.ylabel("Count")
    plt.title(title)

# 各データセットの分布をプロット
# 全体のプロットのサイズを調整

plt.figure(figsize=(12, 4))  # 横幅12インチ、縦幅4インチに調整

# 各データセットの分布を横に並べてプロット
plt.subplot(1, 3, 1)  # 1行3列の1番目
plot_label_distribution(train_counts, "Distribution of Labels in Train Dataset")

plt.subplot(1, 3, 2)  # 1行3列の2番目
plot_label_distribution(val_counts, "Distribution of Labels in Validation Dataset")

plt.subplot(1, 3, 3)  # 1行3列の3番目
plot_label_distribution(test_counts, "Distribution of Labels in Test Dataset")

# グラフの表示
plt.tight_layout()  # レイアウト調整
plt.show()


```

```python
# 必要なライブラリのインポート
import matplotlib.pyplot as plt
from collections import Counter
import numpy as np
from medmnist import INFO
from load_medmnist import load_organmnist_data  # データ読み込み関数をインポート

# OrganAMNISTデータセットのラベル情報を取得
data_flag = 'organamnist'
label_descriptions = INFO[data_flag].get('label', {})

# データセットの準備
train_dataset = load_organmnist_data(data_flag=data_flag, split='train')

# 各ラベルのカウント
labels = [int(label[0]) for _, label in train_dataset]  # ラベルを整数に変換してリストに格納
label_counts = Counter(labels)

# カウントの確認
print("Label Counts:", label_counts)
print("Label Descriptions:", label_descriptions)

# ヒストグラムのプロット
plt.figure(figsize=(10, 6))
plt.bar(
    [label_descriptions.get(i, f"Label {i}") for i in label_counts.keys()],
    label_counts.values()
)
plt.xlabel("Label")
plt.ylabel("Count")
plt.title("Distribution of Labels in OrganMNIST Dataset")
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
import numpy as np
from torchvision import transforms
from medmnist import OrganAMNIST  # 正しいクラスのインポート

DATA_ROOT = "./data"  # データ保存先のパスを指定

def load_medmnist():
    transform = transforms.Compose([transforms.ToTensor()])

    # データのロード
    traindata = OrganAMNIST(root=DATA_ROOT, split='train', transform=transform, download=True)
    testdata = OrganAMNIST(root=DATA_ROOT, split='test', transform=transform, download=True)

    # データとラベルを取得
    X_train, y_train = traindata.imgs, traindata.labels
    X_test, y_test = testdata.imgs, testdata.labels

    # NumPy配列に変換
    X_train = X_train.numpy()
    y_train = y_train.numpy()
    X_test = X_test.numpy()
    y_test = y_test.numpy()

    return X_train, y_train, X_test, y_test
```
